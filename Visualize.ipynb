```python
import os
import pandas as pd
import sklearn
```

```python
fdict = {}
for d in ["./data/val", "./data/train"]:
    for root, dirs, files in os.walk(d):
        path = root.split(os.sep)
        print(path)
        #print((len(path) - 1) * '---', os.path.basename(root))
        for file in files:
            if(not file.split('.')[1]=='npy' ):
                continue
            filepath = os.path.join(root, file)
            print(filepath)
            fid = path[-1]+'_'+file.split('.')[0]
            raw = np.load(filepath, allow_pickle = True)
            print(raw.shape)
            fdict[fid] = {'label':path[-1], 'path':filepath, 'length':raw.shape[0], 'split':d.split('/')[-1]}
```

```python
fdict
```

```python
df = pd.DataFrame(fdict).T
```

```python
df.head()
```

```python
df.groupby('length').size()
```

```python
l = df.groupby(['length', 'label']).size()
```

```python
sklearn.metrics.confusion_matrix()
```
